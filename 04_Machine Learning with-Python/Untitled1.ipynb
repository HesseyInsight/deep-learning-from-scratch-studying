## #0. 들어가기에 앞서

본 책을 집은 이유는 간단하다. 나는 컴퓨터 과학을 제대로 배우고 싶은 대학원 생이지만 어려운 수식으로 가득찬 책들로 공부하기에는 내공이... 많이 부족하다... 이 책은 이런 내게 하늘에서 내리는 단비 같달까...

저자는 프롤로그에서 다음과 같이 말했다. "머신러닝과 인공지능 관련 도서는 많다. 하지만 이러한 책들은 컴퓨터 과학을 전공하는 대학원생을 위한것이며 어려운 수식들로 가득 차있다."

지난 1년간 머신러닝/딥러닝/데이터 과학/빅데이터 등... 최신 기술을 배우기 위해 열심히 무언가를 했다. 안타까운 점은... 대부분의 자료, 서적들이 어려운 수식으로 가득차 있었고 이제 막 공부를 시작하는 내게는 뭐랄까... 오히려 흥미를 감퇴시키는 부작용을 일으켰던것 같다. 봐도 봐도 봐도 모르겠는 내용 뿐이었고, 다시 공부를 하자니 돌아갈 길이 너무 멀었다... 좋은 점은 이 책을 만났다는 것이다.

그렇다고 대충 '아~ 이런거구나'하고 넘기진 않을 것이다. 그럴거면 굳이 '밑바닥부터 시작하는' 공부를 시작하지 않았을 것이다. 이번 학습에서 내가 바라는 점은 사이킷런 핵심 개발자인 저자로부터 머신러닝을 학습하는 하나의 노하우를 전수 받는 것이다.

* 정리하면
    * 이 책을 통해 다양한 머신러닝 알고리즘을 라이브러리를 활용해 구현할 것이다.
    * 연구와 상용 어플리케이션에 사용할 수 있도록 머신러닝을 사용하는 방법을 전수받을 것이다.

한편... 요즘 하루하루 전염병으로 모두가 고생이다. 연구실-집을 오가며 연구자로서 묵묵한 길을 걷고 있는 내게 아직 아무런 일이 일어나지는 않았지만 불안감이 너무 큰 나머지 일이 손에 잘 안잡힌다. 하루빨리 좋은 날이 오기를 기대한다. 훗날 같은 일이 일어났을 때는 내가 가진 기술을 통해 세상에 도움을 주는 사람이 되고 싶다. 모두 화이팅...
